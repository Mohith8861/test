```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../D/TensorFlow_FILES/TensorFlow_FILES/archive/main_dataset.csv")
```

```python
df.head()
```

```python
df.drop(['format','currency','price','isbn','category','old_price'],axis=1,inplace=True)
```

```python
mod_names = []
for i in df['name']:
    x = [a for a in i.lower() if a not in '!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~']
    mod_names.append(''.join(x))
mod_names
```

```python
df['mod_names'] = mod_names
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()

tfidf = vectorizer.fit_transform(df["mod_names"])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import re


def show_image(val):
    return '<a href="{}"><img src="{}" width=50></img></a>'.format(val, val)

def search(query,vectorizer):
    processed = re.sub("[^a-zA-Z0-9 ]", "", query.lower())
    query_vec = vectorizer.transform([query])
    similarity = cosine_similarity(query_vec, tfidf).flatten()
    indices = np.argpartition(similarity, -10)[-10:]
    results = df.iloc[indices]
    results = results.sort_values("book_depository_stars", ascending=False)
    results = results.drop(['mod_names','img_paths'],axis=1)
    
    return results.head(5).style.format({'image': show_image})
```

```python
search("When Breath",vectorizer)
```
